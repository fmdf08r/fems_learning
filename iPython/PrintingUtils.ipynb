```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
exp1_path = "../training/results_1hl_100u_200ep.csv"
exp2_path = "../training/results_1hl_200l_200ep.csv"
exp3_path = "../training/results_1hl_300u_100ep.csv"
exp4_path = "../training/results_1hl_50l_100ep.csv"
exp5_path = "../training/results_2hl_100u_100ep.csv"
exp6_path = "../training/results_2hl_50l_100ep.csv"
exp7_path = "../training/results_3hl_100l_100ep.csv"
```

```python
df1 = pd.read_csv(exp1_path)
df2 = pd.read_csv(exp2_path)
df3 = pd.read_csv(exp3_path)
df4 = pd.read_csv(exp4_path)
df5 = pd.read_csv(exp5_path)
df6 = pd.read_csv(exp6_path)
df7 = pd.read_csv(exp7_path)
```

```python
plt.figure(1,figsize=(15,15))
plt.subplot(221)
plt.title('100u-100ep-4M')
plt.xlabel('# epochs')
plt.ylabel('MSE error')
plt.axis([0, 100, 0, 2])
tst_error = plt.plot(df1['val_loss'].values,color='r',linewidth=2, label='test error')
tr_error = plt.plot(df1['loss'].values,color='b',linewidth=2, label='training error')
plt.legend(loc='upper right')

plt.subplot(222)
plt.title('200u-100ep-4M')
plt.xlabel('# epochs')
plt.ylabel('MSE error')
plt.axis([0, 100, 0, 2])
tst_error = plt.plot(df2['val_loss'].values,color='r',linewidth=2, label='test error')
tr_error = plt.plot(df2['loss'].values,color='b',linewidth=2, label='training error')
plt.legend(loc='upper right')

plt.subplot(223)
plt.title('300u-100ep-4M')
plt.xlabel('# epochs')
plt.ylabel('MSE error')
plt.axis([0, 100, 0, 2])
tst_error = plt.plot(df3['val_loss'].values,color='r',linewidth=2, label='test error')
tr_error = plt.plot(df3['loss'].values,color='b',linewidth=2, label='training error')
plt.legend(loc='upper right')

plt.subplot(224)
plt.title('50u-100ep-4M')
plt.xlabel('# epochs')
plt.ylabel('MSE error')
plt.axis([0, 100, 0, 2])
tst_error = plt.plot(df4['val_loss'].values,color='r',linewidth=2, label='test error')
tr_error = plt.plot(df4['loss'].values,color='b',linewidth=2, label='training error')
plt.legend(loc='upper right')

plt.savefig('../visualizations/test_error_#units.png')
```

```python
plt.figure(figsize=(8,8))
plt.title('100u-[1-2-3]hl-100ep-4M')
plt.xlabel('# epochs')
plt.ylabel('MSE error')
plt.axis([0, 100, 0, 1])
tst_e1 = plt.plot(df1['val_loss'].values,color='r',linewidth=2, label='te 1hl')
tst_e2 = plt.plot(df5['val_loss'].values,color='b',linewidth=2, label='te 2hl')
tst_e3 = plt.plot(df7['val_loss'].values,color='g',linewidth=2, label='te 3hl')
plt.legend(loc='upper right')
plt.savefig('../visualizations/test_error_#layers.png')
```

```python
plt.figure(figsize=(8,8))
plt.title('50vs100u-[1-2]hl-100ep-4M')
plt.xlabel('# epochs')
plt.ylabel('MSE error')
plt.axis([0, 100, 0, 1])
tst_e1 = plt.plot(df4['val_loss'].values,'r',linewidth=2, label='te 50l - 1hl')
tst_e2 = plt.plot(df6['val_loss'].values,'b',linewidth=2, label='te 50l - 2hl')
tst_e3 = plt.plot(df1['val_loss'].values,'g',linewidth=2, label='te 100l - 1hl')
tst_e3 = plt.plot(df5['val_loss'].values,'k',linewidth=2, label='te 100l - 2hl')
plt.legend(loc='upper right')
plt.savefig('../visualizations/test_error_#50vs100.png')
```

```python
plt.figure(figsize=(8,8))
plt.title('[50, 100, 200, 300]u-1hl-100ep-4M')
plt.xlabel('# epochs')
plt.ylabel('MSE error')
plt.axis([0, 100, 0, 0.6])
tst_e1 = plt.plot(df1['val_loss'].values,'r',linewidth=2, label='te 100l')
tst_e2 = plt.plot(df2['val_loss'].values,'b',linewidth=2, label='te 200l')
tst_e3 = plt.plot(df3['val_loss'].values,'g',linewidth=2, label='te 300l')
tst_e3 = plt.plot(df4['val_loss'].values,'k',linewidth=2, label='te 50l')
plt.legend(loc='upper right')
plt.savefig('../visualizations/test_error_#nbr_units.png')
```
